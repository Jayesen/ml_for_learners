```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data = pd.read_csv('raw_data/Student-por.csv')

columns_to_encode = ['school', 'sex', 'Medu','reason','traveltime','studytime', 'freetime', 'higher',
       'failures', 'internet']

columns_not_to_encode = ['age', 'G3']

encoder = OneHotEncoder(sparse=False)

one_hot_encoded = encoder.fit_transform(data[columns_to_encode])

encoded_columns = encoder.get_feature_names_out(columns_to_encode)

encoded_df = pd.DataFrame(one_hot_encoded, columns=encoded_columns)

final_df = pd.concat([encoded_df, data['age'], data['G3']], axis=1)

print(final_df)

```
