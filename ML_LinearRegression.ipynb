```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student-por.csv')
df
```

```python
df_dropped = df[['sex', 'age', 'Medu', 'reason', 'traveltime', 'studytime', 'freetime', 'higher',
       'failures', 'internet', 'G3']]
df_dropped
```

```python
#Pre-processing do a pipeline 

#From string to numeric
df_dropped['sex2'] = df['sex'].replace({'F': 0, 'M': 1})
df_dropped['school2'] = df['school'].replace({'GP': 0, 'MS': 1})
df_dropped['reason2'] = df['reason'].replace({'course': 1, 'home': 2, 'other': 3, 'reputation': 4})
df_dropped['internet2'] = df['internet'].replace({'yes': 1, 'no': 0})
df_dropped['higher2'] = df['higher'].replace({'yes': 1, 'no': 0})

#Reassign failures to 0 or 1 
df_dropped.loc[df_dropped['failures'] > 0, 'failures'] = 1

#Reduce traveltime to binary values: less than 15 mins or greater than 15mins
#df_school.loc[df_school['traveltime'] > 1, 'traveltime'] = 2
df_dropped['traveltime'] = df_dropped['traveltime'].apply(lambda x: 1 if x == 1 else 0)

df_dropped = df_dropped[['sex2', 'age', 'school2', 'Medu', 'reason2', 'traveltime', 'studytime', 'freetime', 'higher2',
       'failures', 'internet2', 'G3']]
df_dropped
```

```python
from sklearn.model_selection import train_test_split

#Split the data into train and test
train_data, test_data = train_test_split(df_dropped, test_size=0.3)

#Ready X and Ys
X_train = train_data[['sex2', 'age', 'school2', 'Medu', 'reason2', 'traveltime', 'studytime', 'freetime', 'higher2',
       'failures', 'internet2']]
y_train = train_data[['G3']]

X_test = test_data[['sex2', 'age', 'school2', 'Medu', 'reason2', 'traveltime', 'studytime', 'freetime', 'higher2',
       'failures', 'internet2']]
y_test = test_data[['G3']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Instantiate the model
model = LinearRegression()

# Train the model on the Training data
model.fit(X_train, y_train)

# Score the model on the Test data
model.score(X_test,y_test)
```
